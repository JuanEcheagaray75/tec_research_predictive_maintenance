```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ruptures as rpt
```

```python
# creation of data
n, dim = 500, 1  # number of samples, dimension
n_bkps, sigma = 2, 1  # number of change points, noise standard deviation
signal, bkps = rpt.pw_constant(n, dim, n_bkps, noise_std=sigma)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(np.ravel(signal))
[plt.axvline(bk, c='k') for bk in bkps]
plt.xlabel('t')
plt.ylabel('f(t)')
plt.suptitle('Sample time series', fontsize=14, fontweight='bold')
plt.title(f'Breakpoints at idx: {bkps}')
plt.savefig('pdm/img/sample_breakpoints.png', dpi=300, bbox_inches='tight')
plt.show()
```
